# Working with multi dimensional arrays in Numpy

- The dimension mentioned here is the dimension of the tensor. It is not the same as the number of dimensions in the vector/array.
- For example, [1,2,3,4,5] is a 1D tensor but a vector with 5 dimensions.
- And as we combine the previous dimensional tensors we get the next higher dimensional tensors. For example, if we combine 3 1D tensors of shape (5,) we get a 2D tensor of shape (3,5) and if we combine 4 2D tensors of shape (3,5) we get a 3D tensor of shape (4,3,5) and so on. 

```python
# 0 D array
scalar = np.array('A')
print("0 D array:", scalar)
print("Number of dimensions:", scalar.ndim)
print("Shape of scalar:", scalar.shape)
```

```python
# 1 D array
array_1d = np.array(['A', 'B', 'C'])
print("1 D array:", array_1d)
print("Number of dimensions:", array_1d.ndim)
print("Shape of 1 D array:", array_1d.shape)
```

```python
# 2 D array
array_2d = np.array([['A', 'B', 'C'], ['D', 'E', 'F']])
print("2 D array:", array_2d)
print("Number of dimensions:", array_2d.ndim)
print("Shape of 2 D array:", array_2d.shape)
```

```python
# 3 D array
array_3d = np.array([[['A', 'B','a'], ['C', 'D', 'b']], [['E', 'F', 'c'], ['G', 'H', 'd']]])
print("3 D array:", array_3d)
print("Number of dimensions:", array_3d.ndim)
print("Shape of 3 D array:", array_3d.shape)

#(depth, rows, columns) -> (2, 2, 3) Shape of 3 D array
```

```python
# 4 D array
array_4d = np.array([[[['A', 'B'], ['C', 'D']], [['E', 'F'], ['G', 'H']]], [[['a', 'b'], ['c', 'd']], [['e', 'f'], ['g', 'h']]]])
print("Number of dimensions:", array_4d.ndim)
print("Shape of 4 D array:", array_4d.shape)

#(blocks, depth, rows, columns) -> (2, 2, 2, 2) Shape of 4 D array
# Visualize it like two blocks stacked on top of each other, each block containing two 2x2 matrices of depth of 2 layers.
```

# Accessing elements in multi dimensional arrays

```python
# lets look for the 3D array
print("Element at depth 1, row 2, column 3:")  # Output: 'b'

# Using chain indexing
element = array_3d[0][1][2]
print("Element using chain indexing:", element)

# using multi-dimensional indexing of numpy
element_np = array_3d[0, 1, 2]
print("Element using numpy multi-dimensional indexing:", element_np)
```
