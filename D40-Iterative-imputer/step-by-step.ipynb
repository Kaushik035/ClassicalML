# There are 3 categories of missing data:
1. **Missing Completely at Random (MCAR)**: For certain rows data collect hi nhi hua for any arbitary reason
2. **Missing at Random (MAR)**: Jo data aap collect kar rhe ho, usme logo ne optionally logo ne data nhi bhara (like weight,sex...etc.)
    - In this case the missing value can be filled using other columns (It means there is a relation)
3. **Missing Not at Random (MNAR)**: Jaan bhuj kar data hataya gaya hai. (So no relation of missing values with other columns)

### <span style="color:green">We use MICE (Multiple Imputation by Chained Equations) when we are sure that the data is MAR</span>

## Advantages of MICE:
1. Accurate
## Disadvantages of MICE:
1. Slow (because aap actual mei ek ML algo ka use kar rhe ho to fill the missing values)
2. Need to store training data in production

# Working of MICE:

- It is always implemented on input columns




- Now we need to fill the missing values using MICE



- Now proceed column wise from left to right and fill the missing values using the below steps:


- So for R&D col, Administration and Marketing col will be the input features and R&D col will be the target feature. So we will train a ML (e.g., Linear Regression) model using the rows which have non-missing values in R&D col and then we will use that trained model to predict the missing value in R&D col.




- Now repeat the process iteratively taking the previous iteration's imputed values, till the differce between the imputed values in 2 consecutive iterations is less than a certain threshold value. (i.e., till convergence is achieved)




(So the initial mean value we used to fill the missing value is just a reasonable guess start value)

- Below code snippet is the implementation of MICE:

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df.head()
```

```python
# Step 1 - Impute all missing values with mean of respective col

df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
# 0th Iteration
df0
```

```python
# Remove the col1 imputed value
df1 = df0.copy()

df1.iloc[1,0] = np.nan

df1
```

```python
# Use first 3 rows to build a model and use the last for prediction

X = df1.iloc[[0,2,3,4],1:3]
X
```

```python
y = df1.iloc[[0,2,3,4],0]
y
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```

```python
df1.iloc[1,0] = 23.14
```

```python
df1
```

```python
# Remove the col2 imputed value

df1.iloc[3,1] = np.nan

df1
```

```python
# Use last 3 rows to build a model and use the first for prediction
X = df1.iloc[[0,1,2,4],[0,2]]
X
```

```python
y = df1.iloc[[0,1,2,4],1]
y
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[3,[0,2]].values.reshape(1,2))
```

```python
df1.iloc[3,1] = 11.06
```

```python
df1
```

```python
# Remove the col3 imputed value
df1.iloc[4,-1] = np.nan

df1
```

```python
# Use last 3 rows to build a model and use the first for prediction
X = df1.iloc[0:4,0:2]
X
```

```python
y = df1.iloc[0:4,-1]
y
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[4,0:2].values.reshape(1,2))
```

```python
df1.iloc[4,-1] = 31.56
```

```python
# After 1st Iteration
df1
```

```python
# Subtract 0th iteration from 1st iteration

df1 - df0
```

```python
df2 = df1.copy()

df2.iloc[1,0] = np.nan

df2
```

```python
X = df2.iloc[[0,2,3,4],1:3]
y = df2.iloc[[0,2,3,4],0]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df2.iloc[1,1:].values.reshape(1,2))
```

```python
df2.iloc[1,0] = 23.78
```

```python
df2.iloc[3,1] = np.nan
X = df2.iloc[[0,1,2,4],[0,2]]
y = df2.iloc[[0,1,2,4],1]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df2.iloc[3,[0,2]].values.reshape(1,2))
```

```python
df2.iloc[3,1] = 11.22
```

```python
df2.iloc[4,-1] = np.nan

X = df2.iloc[0:4,0:2]
y = df2.iloc[0:4,-1]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df2.iloc[4,0:2].values.reshape(1,2))
```

```python
df2.iloc[4,-1] = 31.56
```

```python
df2
```

```python
df2 - df1
```

```python
df3 = df2.copy()

df3.iloc[1,0] = np.nan

df3
```

```python
X = df3.iloc[[0,2,3,4],1:3]
y = df3.iloc[[0,2,3,4],0]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df3.iloc[1,1:].values.reshape(1,2))
```

```python
df3.iloc[1,0] = 24.57
```

```python
df3.iloc[3,1] = np.nan
X = df3.iloc[[0,1,2,4],[0,2]]
y = df3.iloc[[0,1,2,4],1]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df3.iloc[3,[0,2]].values.reshape(1,2))
```

```python
df3.iloc[3,1] = 11.37
```

```python
df3.iloc[4,-1] = np.nan

X = df3.iloc[0:4,0:2]
y = df3.iloc[0:4,-1]

lr = LinearRegression()
lr.fit(X,y)
lr.predict(df3.iloc[4,0:2].values.reshape(1,2))
```

```python
df3.iloc[4,-1] = 45.53
```

```python
df2.iloc[3,1] = 11.22
```

```python
df3
```

```python
df3 - df2
```
