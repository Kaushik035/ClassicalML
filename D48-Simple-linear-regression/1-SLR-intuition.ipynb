# Simple Linear Regression

- 1 input column and 1 output column
- Eg. Based on CGPA, predict the package of a student

## Steps once you have the data:
1. Plot the data



- On plotting, you can see the data is sort of linear
- But its not completely linear (lie in a straight line), it is because in real life there are many real life factors that affect the output, jisko hum mathematically samaj nahi paate (like interview kharab ho gaya) and these are factors are called stochastic errors(jisko hum mathematically samaj nahi paate)


- If it was completely linear, then all the points would have lied on the straight line, so would just need to find the slope and intercept of the line, but since there are stochastic errors, we need to find the line of best fit (line that best fits the data)



Best fit line is the line that minimizes the distance between the line and the data points, so in the pic 1 is the best fit line and not 2

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

## 📌 Pandas, NumPy & ML Shapes (Quick Summary)

## 🔹 Series vs DataFrame
- **Series (1D)** → single column/row  
  Shape: `(n,)`
- **DataFrame (2D)** → table (rows × columns)  
  Shape: `(n, m)`

---

## 🔹 iloc (Indexing)
- `df.iloc[0]` → returns **Series (1D)**
- `df.iloc[[0]]` → returns **DataFrame (2D)** ✅

---

## 🔹 1D vs 2D Arrays
- 1D → `[1,2,3]` → `(n,)`
- 2D → `[[1,2,3]]` or `[[1],[2],[3]]` → `(n,m)`

---

## 🔹 Key Rule
> If shape = `(n,)` → 1D  
> If shape = `(n,m)` → 2D  

---


```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

# Intuition



- Mathematical intuition is we need to find the slope (m) and intercept (b) of the line that best fits the data, so we can write the equation of the line as `y = mx + b`

## <span style="color:orange">But what is the human interpretation of 'm' and 'b'?</span>

- `m` is `weightage` of the feature, it tells us how much the output will change for a unit change in the input
- `b` is the `offset` , suppose the input col was experiece instead of CGPA, so for a person with 0 years of experience, what would be the expected salary, so that is the offset otherwise it would have been 0
