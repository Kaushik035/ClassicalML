# Arbitary value imputation



- Replace missing values with an arbitrary value (e.g. -999, 999, 0, etc.)
- The whole point of this is to create distinction between the missing values and the non-missing
- Mostly used in categorical variables, but can be used in numerical variables as well (e.g. age = 99,-1, etc.)

# When to use arbitrary value imputation?

- <span style="color:orange">When data is not missing at random</span>

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
````

````python
df = pd.read_csv('titanic_toy.csv')
````

````python
df.head()
````

````python
df.isnull().mean()*100 # percentage of missing values in each column
````

````python
X = df.drop(columns=['Survived'])
y = df['Survived']
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
````

````python
X_train['Age_99'] = X_train['Age'].fillna(99)
X_train['Age_minus1'] = X_train['Age'].fillna(-1)

X_train['Fare_999'] = X_train['Fare'].fillna(999)
X_train['Fare_minus1'] = X_train['Fare'].fillna(-1)
````

````python
print('Original Age variable variance: ', X_train['Age'].var())
print('Age Variance after 99 wala imputation: ', X_train['Age_99'].var())
print('Age Variance after -1 wala imputation: ', X_train['Age_minus1'].var())

print('Original Fare variable variance: ', X_train['Fare'].var())
print('Fare Variance after 999 wala imputation: ', X_train['Fare_999'].var())
print('Fare Variance after -1 wala imputation: ', X_train['Fare_minus1'].var())
````

- See bohot difference aa rha h, and this techinque is used for this purpose only, to create distinction between missing and non-missing values

````python
fig = plt.figure()
ax = fig.add_subplot(111)

# original variable distribution
X_train['Age'].plot(kind='kde', ax=ax)

# variable imputed with the median
X_train['Age_99'].plot(kind='kde', ax=ax, color='red')

# variable imputed with the mean
X_train['Age_minus1'].plot(kind='kde', ax=ax, color='green')

# add legends
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)

# original variable distribution
X_train['Fare'].plot(kind='kde', ax=ax)

# variable imputed with the median
X_train['Fare_999'].plot(kind='kde', ax=ax, color='red')

# variable imputed with the mean
X_train['Fare_minus1'].plot(kind='kde', ax=ax, color='green')

# add legends
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
````

````python
trf = ColumnTransformer([
    ('imputer1',imputer1,['Age']),
    ('imputer2',imputer2,['Fare'])
],remainder='passthrough')
````

````python
trf.fit(X_train)
````

# End of Distribution Imputation



- Extension of arbitrary value imputation, where we replace the missing values with an arbitrary value that is outside the distribution of the variable/end points of the variable (i.e. we fill the missing values with outliers)

1. <span style="color:orange">For Normally distributed variables, we can use</span> ```mean + 3*std or mean - 3*std ``` <span style="color:orange">as the arbitrary value for imputation</span>





```Filling missing values with mean ± 3*std can introduce extreme values (almost outliers).```

2. <span style="color:orange">For skewed distributed variables, we can use</span> ```Q1 - 1.5 * IQR OR Q3 + 1.5 * IQR, IQR = Q3 - Q1``` <span style="color:orange">as the arbitrary value for imputation</span>
