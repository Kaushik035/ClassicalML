# Mathematical Intuition of Naive Bayes Classifier



- We have to find: $P(C_k \mid X)$







- Now let's take a **Naive(Simple)** assumption: All features are independent of each other

- It means: $X_i$ is independent of $X_j$ for $i \neq j$



# Final Formula:

$$
P(C_k \mid X) \propto P(C_k) \prod_{i=1}^{n} P(X_i \mid C_k)
$$

# Prediction:

- Calculate $P(C_k \mid X)$ for all classes and predict the class with the highest value

- It is called **Maximum A Posteriori (MAP)** decision rule
