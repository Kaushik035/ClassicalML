# Handling Numerical Data in Naive Bayes Classifier



- As you can see the input of H(height) is not present in the training data, so we cannot calculate the probabilities for height being 185, provided gender is Male

- So we make a strong assumption that the heights of males follow a normal distribution 

- And we can calculate the mean and standard deviation of the heights of males in the training data

- Then we can use the normal distribution to calculate the probability of height being 185 given gender is Male

- Formula of findin g the probability of a value x in a normal distribution with mean μ and standard deviation σ is given by:

# $$P(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

- Similarly we will do it for weights of males, heights and weights of females as well

# But what if our assumption of normal distribution is wrong (the data can follow other distributions as well)?

- The best thing is to plot the data and see which distribution it follows, and then use that distribution Naive Bayes to calculate the probabilities

- The above method is called Gaussian Naive Bayes, and it is used when the features are continuous and we assume that they follow a normal distribution

- Similarly we can have other types of Naive Bayes classifiers for different types of data and distributions, such as Multinomial Naive Bayes for discrete data and Bernoulli Naive Bayes for binary data.
