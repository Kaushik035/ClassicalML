#### Question: There are 3 factories producing markers (M1, M2, M3). M1 produces 20% of the total markers, M2 produces 30%, and M3 produces 50%. The defect rates for the markers are 5% for M1, 3% for M2, and 1% for M3. If a marker is selected at random and found to be defective, what is the probability that it was produced by M3?

- Have to find P(M3|D) where M3 is the event that the marker was produced by factory M3 and D is the event that the marker is defective.
