```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('models/ohe_sex.pkl','rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl','rb'))
clf = pickle.load(open('models/clf.pkl','rb'))
```

```python
# Assume user input
# Pclass/gender/age/SibSp/Parch/Fare/Embarked
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1,7)
# .reshape(1,7) is used to convert the 1D array to 2D array because the model expects the input in 2D array format
# where 1 is the number of samples and 7 is the number of features
```

```python
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

- In the below concetanation step we  combined the transformed numerical and categorical features to create the final transformed training data in the same order as the original training data.

```python
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```

- So we did the same series of steps while predicting as well. As we did while training the model
- So if sometime you change the preprocessing steps in the training phase then you also have to change the same steps in the prediction phase in production. This is a major drawback of not using pipelines. If we were using pipelines then we would have done all these steps in a single line of code and we would not have to worry about changing the same steps in the prediction phase when we change the preprocessing steps in the training phase.
