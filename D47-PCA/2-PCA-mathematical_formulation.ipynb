- PCA is just trying to reduce the dimensionality of our dataset, while keeping the essence of the data in the lower dimension as well

# <span style="color:orange">What is the mathematically objective function that PCA solves?</span>

- So in the below 2D dataset image, we want to reduce the dimension to 1D, keeping the variance as same as it is in 2D



- Let's take example of a single point:



- projection of the point on the line is given by: $x' = x \cdot u$, where $u$ is the unit vector along the line
- So for all the points, we can write: $X' = X \cdot u$, where $X$ is the original dataset and $X'$ is the projected dataset

### <span style="color:orange">Now which unit vector to choose?</span>

- We want to choose the unit vector such that it maximizes the variance of the projected data



- Variance of the projected data is given by:

$$
\text{Var}(X') = \frac{1}{n} \sum_{i=1}^{n} \left(u^T x_i - u^T \bar{x} \right)^2
$$

- This can be rewritten as:

$$
\text{Var}(X') = u^T \left( \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(x_i - \bar{x})^T \right) u
$$

- Let the covariance matrix be:

$$
\Sigma = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(x_i - \bar{x})^T
$$

- Then,

$$
\text{Var}(X') = u^T \Sigma u
$$

- So we need to find the unit vector $u$ that maximizes the variance of the projected data

### What do you mean by the variance of the projected data?

- The variance of the projected data is a measure of how spread out the projected data points are along the line defined by the unit vector $u$.



- So, variance is the Mathematical objective function that PCA wants to maximize.So it is now an optimization problem where we want to find the unit vector $u$ that maximizes the variance of the projected data



# <span style="color:orange">Covariance and Covariance Matrix</span>



- Variance is a one axis measure of spread, while covariance tells us how two variables are related to each other. It can be positive, negative or zero.



- For the above data in the pic, mean and variance of both the variables are same, but covariance is negative for the left plot and positive for the right plot. So covariance gives us more information about the relationship between the variables.
- +ve covariance means, as one variable increases, the other variable also increases. -ve covariance means, as one variable increases, the other variable decreases. Zero covariance means, there is no relationship between the variables.
- Covariance is a stastical qty that tells us how axix are related to each other.
- Correlation is a normalized version of covariance, which gives us the strength of the relationship between the variables. It ranges from -1 to 1.

## COVARIANCE MATRIX






Let $x_i \in \mathbb{R}^d,\; i = 1,2,\dots,n$ denote the data points.

---

### Mean
$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

---

### Covariance Matrix
$$
\Sigma = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(x_i - \bar{x})^T
$$

---

### Expanded Form
$$
\Sigma =
\begin{bmatrix}
\mathrm{Var}(x_1) & \mathrm{Cov}(x_1,x_2) & \cdots & \mathrm{Cov}(x_1,x_d) \\
\mathrm{Cov}(x_2,x_1) & \mathrm{Var}(x_2) & \cdots & \mathrm{Cov}(x_2,x_d) \\
\vdots & \vdots & \ddots & \vdots \\
\mathrm{Cov}(x_d,x_1) & \mathrm{Cov}(x_d,x_2) & \cdots & \mathrm{Var}(x_d)
\end{bmatrix}
$$

---

### Key Insight
$$
\boxed{
\text{Variance measures spread along one axis, \quad Covariance measures joint variation}
}
$$

---

### What is the speciality of the covariance matrix?

1. It tells us the spread of the data along each axis (as diagonal elements are variances).
2. It tells us how the axes are related to each other (as off-diagonal elements are covariances).
3. It is symmetric and positive semi-definite, which allows us to perform eigen decomposition to find the principal components in PCA.

- Symmetric means, $\Sigma_{ij} = \Sigma_{ji}$, and positive semi-definite means, for any non-zero vector $z$, we have $z^T \Sigma z \geq 0$.



# Eigen decomposition of Covariance matrix

- Matrix multiplication is linear transformation of all the vectors in the space. (The magnitude and direction of the vectors changes)





## <span style="color:orange">What are Eigenvectors ?</span>

- <span style="color:lightblue">Eigenvectors are the special vectors that do not change their direction after the linear transformation, they only change their magnitude.</span>

- If magnitude is  greater than 1, then the vector is stretched, if magnitude is less than 1, then the vector is compressed.

- If magnitude is positive, then the vector is in the same direction, if magnitude is negative, then the vector is in the opposite direction. (Though still its eigenvector, just in the opposite direction)

- For a n-dimensional space/coordinate system, we can have at most n linearly independent eigenvectors.

## Eigenvectors — Quick Revision

### Key Idea
In an $n$-dimensional space, you can have **at most $n$** linearly independent eigenvectors.

---

### Why?
- Eigenvectors are just directions in space  
- In $ \mathbb{R}^n $, we can have at most $n$ independent directions  

---

### Two Important Cases

#### 1. General Matrices
- May have **less than $n$** independent eigenvectors  
- This happens when directions “collapse”  
- Such matrices are called **defective**

---

#### 2. Symmetric Matrices (e.g., Covariance Matrix)
- Always have **exactly $n$** independent eigenvectors  
- Eigenvectors are **orthogonal** (perpendicular)

---

### Key Result
$$
\text{Number of independent eigenvectors} \le n
$$

---

### Intuition
- Think of eigenvectors as **important directions of data**
- Maximum possible directions = dimension of space  
- Symmetric matrices give **all possible clean directions**

---

# What are Eigenvalues?

- Eigenvalues are the magnitudes of stretching/compressing that occur along the eigenvector direction after the linear transformation.

- There can be multiple vectors corresponding to the same eigenvalue, but each eigenvector will have its own eigenvalue that indicates how much it is stretched or compressed.





 <span style="color:orange"> **So,using Rayleigh quotient, we maximised the variance (Objective Function) and found that the eigenvector corresponding to the largest eigenvalue of the COVARIANCE matrix gives us the direction of maximum variance in the data. This is the first principal component in PCA. The second principal component is given by the eigenvector corresponding to the second largest eigenvalue, and so on.** </span>

So, steps are:
1. Compute the covariance matrix of the data.
2. Perform eigen decomposition to find eigenvalues and eigenvectors.
3. Sort the eigenvectors by their corresponding eigenvalues in descending order.
4. The vector corresponding to the largest eigenvalue is the first principal component, the vector corresponding to the second largest eigenvalue is the second principal component, and so on.

# <span style="color:orange">Step by step solution</span>

1. Mean center the data by subtracting the mean from each data point along all dimensions/columns/features. (Without this step also PCA works, but it is a good practice to mean center the data before applying PCA for better results)



2. Compute the covariance matrix of the mean-centered data. (fornd using np.cov)



3. Find the eigenvalues and eigenvectors of the covariance matrix. (fornd using np.linalg.eig)



 - Now you want to go to 1D from 3D then just take PC1, if you want to go to 2D from 3D then take PC1 and PC2, and so on.

4. Now that we got the principal components, now we just need to project the original data onto the principal components to get the reduced dimensionality data.


- If earlier dimension of data set was (1000,4){including the target variable} and we want to reduce it to 1D, then after projection the dimension of the data set will be (1000,2)



For 2D projection, we will take the first 2 principal components and project the data onto them. 




## Projection: 3D → 1D (Matrix Form with Dimensions)

### Original Data

Let:
$$
X \in \mathbb{R}^{1000 \times 3}
$$

- $1000$ rows → data points  
- $3$ columns → features  

We can write:
$$
X =
\begin{bmatrix}
- & x_1^T & - \\
- & x_2^T & - \\
& \vdots & \\
- & x_{1000}^T & -
\end{bmatrix}
$$

where each:
$$
x_i \in \mathbb{R}^{3 \times 1}
$$

---

### Principal Direction

Let:
$$
u_1 \in \mathbb{R}^{3 \times 1}
$$

---

### Step 1: Projection of a Single Point

$$
z_i = u_1^T x_i
$$

Dimensions:
$$
(1 \times 3)(3 \times 1) = (1 \times 1)
$$

So, each $z_i$ is a **scalar**

---

### Step 2: Projection of Entire Dataset

$$
Z = X u_1
$$

Dimensions:
$$
(1000 \times 3)(3 \times 1) = (1000 \times 1)
$$

---

### Step 3: Expanded View

$$
Z =
\begin{bmatrix}
x_1^T u_1 \\
x_2^T u_1 \\
\vdots \\
x_{1000}^T u_1
\end{bmatrix}
$$

Each entry:
$$
z_i = u_1^T x_i
$$

---

### Final Result

$$
X \in \mathbb{R}^{1000 \times 3}
\quad \longrightarrow \quad
Z \in \mathbb{R}^{1000 \times 1}
$$

---

### Key Insight

- Rows remain same → number of data points  
- Columns reduce → number of features  

Projection = **collapse each 3D point into 1 number**

---

# If you are wondering about the equivalence of the two forms of projection:

- Dont worry about the order of multiplication, both forms are equivalent and give the same result its just a matter of notation and perspective.

- For a single point: $z_i = u_1^T x_i$

- For the dataset: $Z = X u_1$

- Both are equivalent:
  $$
  u_1^T x_i = x_i^T u_1
  $$

- Dataset form applies the same projection to all points simultaneously.
