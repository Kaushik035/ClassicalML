# PCA (Principal Component Analysis)

- It's an unsupervised method for dimensionality reduction (Isme data mei bss input hota hai output nhi hota).
- It's an old complex reliable technique
- It's a feature extraction technique jisse COD (Curse of Dimensionality) reduce hota hai by decreasing the number of features.

## Analogy: 

 The photographer trying to capture the 3D football match into the best 2D image possible. The photographer will try to capture the most important features of the match (like players, ball, etc.) while ignoring less important details (like the crowd, background, etc.). 

Similarly, PCA is a technique which transforms higher dimensional data to lower dimensional data while keeping the essence of the data intact.

## Benefits of PCA:
1. Faster execution of algorithms (as we have reduced the number of features, so training data is smaller)
2. Visualization of data in 2D or 3D

# Geometric Intuition of PCA

- It's a house data in the picture
- We saw that there are 2 ways to reduce COD: Feature Selection and Feature Extraction, and the most important technique for feature extraction is PCA.
- But to get geometric intuition of **PCA** we need to understand little bit about **feature selection** first.

## Feature selection:
- Iska basic funda hai ki if you have 5 columns in your data, you will select the most important 2 or 3 columns and removev  the rest.



- So, from our domain knowledge if we have to remove 1col then we wil remove "Number of grocery shops"
- But in actual work, datasets are much more complex and we don't have domain knowledge, so we can't just select features based on our intuition. We need a systematic way to select features, so for that let me show you a mathematical trick:

### How to find which column is more important than the other? (even agar aapko koi idea naa ho data ko leke)

- Take both the features and plot them each on x and y axis and now draw projection of the data points on their respective axes. Now, we can see that the projection on the x-axis is much more spread out than the projection on the y-axis. So, we can say that the x-axis is more important than the y-axis and we can select the x-axis as our feature and remove the y-axis.



- We can connect using variance (Jis axis mei variance zada hoga uss feature ko lenge) and this is how feature selection works, but it has a major drawback that we are losing a lot of information by just selecting one feature and removing the other feature.

### But what if variance similar hai dono features mei? (yaani dono features important hai)



- For this case we can't just eliminate any feature, bcoz both numbe of  washrooms and no. of rooms will decide the price of the house
- Both the variance are similar and no. of washrooms depends on the number of rooms, so we can't just select one feature and remove the other feature.
- So now even mathematically decide which feature to select....This is where feature extraction helps us

# How will PCA help us in this case?

### 1. Example Intution:
- We have 2 features: number of rooms and number of washrooms
- So if we go to a broker, and he suggests that as both the colums are imp, so we can rather take ```size of the house``` as a feature which is a combination of both the features.
- And this is exactly what PCA does, it creates a new feature which is a combination of the original features and take some subset of the new features as the final features for our model.



But what if we don't have a broker to suggest us the feature, how will we find the new feature? So lets see the geometric intuition of PCA to understand how it works.

### 2. Geometric Intuition:

- PCA will find a new axis (or direction) in the feature space which is the direction of maximum variance in the data. This new axis is called the principal component. The first principal component will be the direction of maximum variance, the second principal component will be the direction of maximum variance orthogonal to the first principal component, and so on.



- So in the picture , we got room' as PC1 and washroom' as PC2. So, we can see that the variance along the PC1 is much more than the variance along the PC2, so we can select PC1 as our new feature and remove PC2.

- <span style="color:orange">Number of PC <= Number of Original Features (always)</span> (we can have at max as many PC as the number of original features, but we can select less than that)

# What is Variance and why is it important?



- MEAN ka kaam hai central tendency ko capture karna(means humara center kaha hai), aur VARIANCE ka kaam hai data ke spread ko capture karna (hrr point mean se kitna dur hai).
- In the pic, you can tell whats the difference between the 2 data using MEAN, but VARIANCE will tell you how spread out the data is around the mean.



- See variance of 2nd data is 4 times the variance of 1st data, so we can say that the 2nd data is more spread out than the 1st data.

### <span style="color:orange">Spread is not same as variance (Spread is proportional to variance)</span>

### Mean Absolute deviation (MAD)
- MAD is the average of the absolute deviations from the mean. It is a measure of the spread of the data around the mean.

### <span style="color:orange">But why do we use variance instead of MAD for PCA as both are measures of spread?</span>
- Because MAD is a modulus function and it is not differentiable at 0, so it troubles the optimization process. On the other hand, variance is a quadratic function and it is differentiable everywhere, so we can use it for optimization.

# Why is Variance important?
- If we do PCA and bring the 2D data to 1D



- If we take projections in y-axix, see the actual distance b/w red and green point is very less, however if we take projection in x-axis, the distance b/w red and green point is similar to the actual distance b/w red and green point

### <span style="color:orange">So higher dimension se jb lower dimension mein convert kr rhe ho toh data points kaa aapas mei relation gadbad na ho, iss liye variance MAXIMIZE karna hota hai</span>

- <span style="color:lightblue">If we revert to the intial example of the photographer, if the photographer takes photo from the side where players are spreaded out(high variance), then all the players will be visible in the photo, but if the photographer takes photo from the side where players are close to each other(low variance), then some players will be hidden behind others and we won't be able to see them in the photo. So, we want to take the photo from the side where players are spreaded out(high variance) to capture all the important features of the match.</span>
