# Why Lasso Regression creates sparsity ?

- **Sparsity** means that some of the coefficients are exactly zero, which means that those features are not contributing to the model and we can ignore them. This is a desirable property in many cases as it helps in feature selection and also makes the model more interpretable.



- In case of Ridge regression, you can see above that on increasing the value of $\lambda$ , the coefficients are shrinking but they are not becoming zero, they are just becoming very small but not exactly zero. So we can say that Ridge regression does not create sparsity in the model.



- In case of Lasso regression, you can see above that on increasing the value of $\lambda$ , the coefficients are shrinking and after certain value of $\lambda$ , some of the coefficients become exactly zero. So we can say that Lasso regression creates sparsity in the model.

- So the question is :

## Kyu Lasso k case mei sparsity aati hai and Ridge k case mei nahi aati hai ?

- To understand this, we will understand the math using 1 input feature and 1 output feature

- So for linear regression:

    - ### $b$ = $\bar{y} - m\bar{x}$
    
    - ### $m$ = $\frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$


- For Ridge regression:

    - ### $b_{ridge}$ = $\bar{y} - m\bar{x}$
    
    - ### $m_{ridge}$ = $\frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2 + \lambda}$


- For Lasso regression:

    - ### $b_{lasso}$ = $\bar{y} - m\bar{x}$

    

    - we took $2 \lambda |m|$ for simpler derivation

    

    

- So humare iss formulae mei hi chupa hai ki Lasso regression kese coefficients ko 0 kr paata hai and kyu ridge regression nahi kr paata hai.



- So in the pic we took $YX = 100$ and $X^2 = $50$ (just random values to demonstrate the concept)


$$
m_{lasso} = \frac{100 - \lambda}{50} 
$$ 

So, as we increase the value of $\lambda$ , the value of $m_{lasso}$ is decreasing and after value of $\lambda = 100$ , the value of $m_{lasso}$ becomes zero and after that it becomes negative but we are taking the $m > 0$ case, so for $\lambda > 100$, we have to take the other form of the formula which is 
$$
m_{lasso} = \frac{100 + \lambda}{50}
$$
 
and as we can see, as we are increasing the value of $\lambda$ , the value of $m_{lasso}$ is increasing

# <span style="color:orange">Summary</span>

1. Lasso mei $\lambda$ numerator mei add hota hai, isliye jab $\lambda$ ki value badhati hai to numerator zero ho jata hai aur slope bhi zero ho jata hai, isliye Lasso regression mei sparsity aati hai.

2. Ridge mei $\lambda$ denominator mei add hota hai, isliye jab $\lambda$ ki value badhati hai to numerator zero nahi hota hai aur slope bhi zero nahi hota hai, isliye Ridge regression mei sparsity nahi aati hai.

- <span style="color:blue">We can also visualize this using the L1 and L2 norm ball and the contours of the loss function, where in order to minimize the loss function we have to minimize both the MSE loss and the regularization term, so the answer will be at the point where the contour of the loss function touches the L1 norm ball for Lasso regression and the L2 norm ball for Ridge regression. </span>

    - As for the case of Lasso regression, the the loss function in 3D will be like parabolic bowl (in 2D contour its elliptical) and the L1 norm ball will be like a diamond shape with corners on the axes, so the contour of the loss function will touch the L1 norm ball at the corners which means that some of the coefficients will be zero, which is the sparsity.

    - Where as in Ridge regression, the the loss function in 3D will be like parabolic bowl (in 2D contour its elliptical) and the L2 norm ball will be like a circle, so the contour of the loss function will touch the L2 norm ball at the point where both coefficients are non-zero, which means that no coefficient will be zero, which is the non-sparsity.
