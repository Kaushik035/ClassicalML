# Why ROC Curve?

- There are 2 major reasons why we use ROC curve:
    1. Threshold selection
    2. It is used for model comparison

# 1. Threshold selection

- It is related to classification (specifically binary classification)

- If you see the inner workings of a classification model, it gives us the probability of the input belonging to a certain class. For example, if we have a binary classification model that classifies whether an email is spam or not spam, it will give us the probability of the email being spam.

- Then we set a threshold (like 0.5) to classify the email as spam or not spam. If the probability is greater than 0.5, we classify it as spam, otherwise we classify it as not spam.



- But this 0.5 threshold is not always the best threshold for all the problems. For example, if we are trying to detect a rare disease, we might want to set a lower threshold (like 0.1) to classify more people as having the disease, even if it means that we will have more false positives.

- But we don't know what the best threshold is for our problem. This is where ROC curve comes in. It helps us to find the best threshold for our problem.

# True Positive Rate (TPR)

$$
\text{TPR} = \frac{\text{TP}}{\text{TP} + \text{FN}}
$$

- TPR is also known as Recall. It is the proportion of actual positives that are correctly identified as such.

- Suppose we are making a spam classifier, and we have 200 emails, out of which 100 are spam and 100 are not spam. If our model correctly identifies 80 emails as spam (TP) and incorrectly identifies 20 emails as not spam (FN), then the TPR will be:
$$
\text{TPR} = \frac{80}{80 + 20} = \frac{80}{100} = 0.8
$$



- This formula is telling us out of all the spams, how many we correctly identified as spams.

- ### Always relate TPR to the word **BENEFIT**

- So more the TPR, more the benefit, because we are correctly identifying more spams. (thats the work of our model, to identify spams correctly)

- ### <span style="color:orange"> It simply means, that system jis cheez k liye bana tha, us cheez ko kitna acha perform kar raha hai. </span>

- We want to maximize TPR, because we want solve our problem (which is to identify spams correctly) as best as possible.

- For, 100 spam mails, TPR = 1 when FN = 0, and TPR = 0 when TP = 0.

# False Positive Rate (FPR)

$$
\text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}}
$$

- Related to the word **COST**

- It means out of all the not spams, how many we incorrectly identified as spams. (which is the cost of our model, because we are incorrectly identifying not spams as spams)

- ### <span style="color:orange"> It simply means, that humara system kaam krne k liye kitna cost karwata hai </span>

- So, we want to minimize FPR, because we want to solve our problem (which is to identify spams correctly) with as less cost as possible.

- For, 100 not spam mails, FPR = 0 when FP = 0 and TN becomes 100, and FPR = 1 when TN = 0 and FP becomes 100.

# ROC (Receiver Operating Characteristic) Curve

- It is a plot of TPR vs FPR at different thresholds.

- It means it is a b/w **Benefit** and **Cost**, how much benefit we are getting for a certain cost.

- So we want to reach a point where we are getting maximum benefit for minimum cost (which is the top left corner of the plot).

  

- **Case 1:** For a low value of threshold (say 0.1), we will have high TPR and high FPR, because we are classifying more emails as spam, which means we are correctly identifying more spams (high TPR) but also incorrectly identifying more not spams as spams (high FPR).



- **Case 2:** For a high value of threshold (say 0.99), we will have low TPR and low FPR, because we are classifying less emails as spam, which means we are correctly identifying less spams (low TPR) but also incorrectly identifying less not spams as spams (low FPR).



- Now you might feel there is direct proportional **(linear)** relationship between TPR and FPR. But its not linear actually, it is more like a curve

- Let's say we decresed the threshold from 0.99 to 0.8, then there will be a increase in TP , but there will be very less increase or no increase to FP (as for an 'not spam' email to be classified as 'spam', it has to have a probability greater than 0.8, which is very less likely {as those are mostly the points with probabibility near 0.4-0.5} ), so there will be a significant increase in TPR but very less increase in FPR.

- So graph will move more in the vertical direction (TPR) than in the horizontal direction (FPR), which is why we get a curve instead of a straight line.



- If we increase the threshold from 0.1 to 0.2, then there will be very low  decrease or no decrease in TP, but FP will reduce than before, so there will be a decrease in FPR but very less decrease in TPR.



- And we take the point of the curve which is closest to the top left corner, because that point will give us the maximum benefit for minimum cost.
