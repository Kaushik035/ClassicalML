# Handling mixed data



- you can see in the above image we can split the data into two parts, one is the numerical data and the other is the categorical data.

- Sometimes in a col we have data like this:

| col  |
|------|
| 1    |
| 2    |
| A    |
| 5    |
| K    |

In this case, make 2 columns, one for the numerical data and the other for the categorical data.

| col_num | col_cat |
|---------|---------|
| 1       | NA      |
| 2       | NA      |
| NA      | A       |
| 5       | NA      |
| NA      | K       |

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with') # A: Alone
```

- You can use the below code:

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

- So now you can fill the NA values in the col_num with 0 and in the col_cat with 'NA' or 'Unknown' or 'Missing'.

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
