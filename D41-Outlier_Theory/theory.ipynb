# When is outlier dangerous?

- Outlier hai ye pata chal jayega aaram se, but uske saath krna kya hai ye difficult h sochne mei (like usko remove krna hai ya nahi, usko replace krna hai ya nahi, etc.)

- For eg. in the previous pic, two students studied for less hours still more mark... so if we remove those two students, then the regression line will be more accurate for the rest of the students.
But if we include one more featur like "IQ" of the students, then those two students might have higher IQ, so they are not outliers anymore, and we can keep them in the dataset.

- In Anamoly detection, outliers are the main point of interest, so we cannot remove them.

- Suppose in some date age = 300, which is not possible, so we can safely remove that data point as it is a clear outlier and it will harm the model.


# Effect of outliers on ML algorithms

- Below are some of the algorithms which are affected by outliers:

    - Linear regression
    - Logistic regression
    - Adaboost
    - Neural Networks

<span style="color:orange">The common thing in all the above algo is that we calculate **weights** in all of them</span>

- So kvhi vhi agar aap ye soch rhe ho ki outliers se model pe asar padega ki nhi,then see if your model is calculating weights or not, if it is calculating weights, then outliers will affect your model.

- Below are some of the algorithms which are not affected by outliers(because ye dataspace ko cut krke decision leta hai, so outliers ka asar nhi hota):

    - Decision Tree
    - Random Forest
    - Xgboost
    - Gradient Boosting

# How to treat outliers?

1. Trimming (outliers ko remove krdo)
    - data becomes less if more outliers are there
    - but its fast and simple
2. Capping (outliers ko replace krdo with some value like outlier cutoff value)
    - data remains the same
    - but the extreme values are capped
3. Treat it like missing value
4. Discreatization/Binning
