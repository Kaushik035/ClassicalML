# What is Feature Scaling?


- we bring dow both input feattures(iq,cgpa) to a similar range, so that both features can contribute equally to the model performance. This is called feature scaling.





- So, the new coloumn age' will now have mean 0 and standard deviation 1. (this is called standardization)

2. is the spread of salary

- First thing in standerdization is that we make mean = 0 for the feature (Mean centering ->as we can see in the pic)


- Dusri chees ye hai ki we make standard deviation = 1 for the feature - For ex, in the pic sd = 5 for the age feature and sd = 6 for the salary feature.

- So, making sd =1, means we are squishing the data points closer to the mean
- also id sd <1 , then  doing standardization (sd =1) would be expanding of data points away from the mean.



## Example of Standardization

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.sample(5)
```

```python
df=df.iloc[:,2:]
```

```python
df.sample(5)
```

### Whenever you do feature scaling(Standardization or normalization), you should always do it after train test split. Why? Because you want to avoid data leakage. If you do feature scaling before train test split, then the mean and standard deviation of the entire dataset will be used to scale the features, which can lead to data leakage. This is because the test set will have information about the training set, which can lead to overfitting.

## Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
# the StandardScaler will standardize features

scaler = StandardScaler() 

# fit the scaler to the train set, it will learn the parameters
# fit krne ka matlab hai ki scaler ko train set ke mean aur std dev ke baare mein batana
scaler.fit(X_train)

# transform train and test sets
# transform ka matlab hai ki, fit se humne mean and variance seekh liya hai, ab hum un parameters ka use karke data ko scale karenge
# using the standardization formula: xi' = (xi - mean) / std deviation
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**VV IMPORTANT: Seekhte hai hum train data se, but transform krte hai dono train and test data ko**

```python
print(scaler.mean_)
print(scaler.var_)
```

```python
X_train
```

```python
X_train_scaled
# X_train was a dataframe, but after scaling it becomes a numpy array
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled.head()
```

```python
np.round(X_train.describe(), 1)
# see before standardization, the mean and std dev were different for each feature(Age and EstimatedSalary)
```

```python
np.round(X_train_scaled.describe(), 1)
```

# Effect of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

- If you notice, there is no change in the shape of the data, but the mean and standard deviation of the features have changed. Now, both features have mean 0 and standard deviation 1, which means they are on the same scale and can contribute equally to the model performance.

- Standardization the actual faida dekhne k lye, see the below Probability density plot

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1, label='Age')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1, label='EstimatedSalary', color='orange')
ax1.legend()

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2, label='Age')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2, label='EstimatedSalary', color='orange')
ax2.legend()

plt.show()
```

- So before standardization, see age is between 0-100, so it peaked,while salary is between 0-150k, so it is more spread out. Dono features ka scale different hai, so they are contributing differently to the model performance.
- After standardization, both features are on the same scale, so they are both contributing equally to the model performance. This is the actual faida of standardization.

# Comparison of Distributions (Individual features)

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

- See in the above plot, distribution remains same after standardization, but the scale of the features has changed.

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

# Why scaling is important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy before scaling:", accuracy_score(y_test, y_pred))
print("Accuracy after scaling:", accuracy_score(y_test, y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
# Decision Tree is not affected by feature scaling, so we get the same accuracy before and after scaling.
#(Atleast we should get the same accuracy, but due to randomness in the algorithm, we might get different accuracies. But in general, we should get the same accuracy.)
```
