# Bias Variance Trade-off


### Bias:

- The inability of a model to truly capture the underlying relationship between the features and the target variable in the **training data** (It means model pattern hi nhi samajh paa rha hai training data ka)



- In the above pic, **M3** can exceptionally capture the underlying relationship between the features and the target variable in the training data, so it has low bias. M1 has the highest bias because it cannot capture the underlying relationship between the features and the target variable in the training data. M2 has a moderate bias because it can capture some of the underlying relationship between the features and the target variable in the training data, but not all of it.

### Variance:

- The sensitivity of the model to small fluctuations in the training data 

- Suppose there are 2 datasets one for training and one for testing. 
    - Suppose we get 10% error on the training data and 100% error on the testing data.
        - This means the variance of the model (100 - 10 = 90%) is very high because the model is very sensitive to small fluctuations in the training data.

    - Suppose we get 10% error on the training data and 15% error on the testing data.
        - This means the variance of the model (15 - 10 = 5%) is very low because the model is not very sensitive to small fluctuations in the training data.



- In M1,M2 both training and test data have similar error, so they have low variance. M3 has a very low error on the training data and a very high error on the test data, so it has high variance.

### Overfitting:

- Jab humara ML model training data pe bohot accha kaam krta hai lekin test data pe accha kaam nhi krta (i.e. training error is very low and testing error is very high, so difference is very high {**like in case of M3**})

### Underfitting:

- The type of ML model jo training data mei hi acche se relationship ko capture nhi kar paa rha (**like M1**)

## <span style="color:orange">We always go for a model with **low bias** and **low variance**</span>

- So this is the trade-off between bias and variance

- So, we have to find the sweet spot between a very simple model(M1) that underfits and a very complex model(M3) that overfits

- And to find the sweet-spot we have 3 popular techniques:

    1. **Regularization**

    2. **Bagging**

    3. **Boosting**
