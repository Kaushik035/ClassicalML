- We learned in the previous lecture that to prevent overfitting, there are 3 main techniques:
    1. Regularization
    2. Bagging
    3. Boosting

# Regularization

- It is a technique using which you induce some added information to the model to reduce overfitting.

- Regularize word se vhi samaj aa rha hai, ki agar kuch jada hai toh use kam krna hai, toh regularization me bhi aap model ke complexity ko kam krte ho, taki wo training data ke noise ko na seekh paye.

- Use it jaha vhi overfitting ka chance hai

- Bohot saare engineers by default isko apne model mei use krte hai, as ise rakhne se koi nuksaan nahi hai

# There are 3 main types of regularization techniques:
    1. L1 Regularization (Lasso Regression)
    2. L2 Regularization (Ridge Regression)
    3. Elastic Net Regularization (Combination of L1 and L2)
