# Effect of learning rate:



- For a fixed number of epochs, if the learning rate is too high, we will be overshooting the optimal value and we will be **diverging** from it. If the learning rate is too low, we will be taking very small steps towards the optimal value and it will take a **long time** to converge to it.

- So, its a hyperparameter that we need to tune it.

# Universality of GD:

- Avhi hum is Linear Regression k saath padh rhe hai, toh aapko esa lag rha hoga ki GD is only for Linear Regression, but actually GD is a very general optimization algorithm that can be used for a wide range of problems, not just Linear Regression.

- We just need slope of the loss function with respect to the parameters we want to optimize, and then we can use GD to update those parameters.

- And loss function is different for different problems, for example in Logistic Regression, we use log loss, in Neural Networks we use cross-entropy loss, etc. But the idea is the same, we calculate the slope of the loss function with respect to the parameters and then we update those parameters using GD.

## <span style="color:orange">The only important thing is this equation of GD:</span>

## $$\theta = \theta - \eta \cdot \text{loss\_slope}$$

- where the slope can be of any loss function
