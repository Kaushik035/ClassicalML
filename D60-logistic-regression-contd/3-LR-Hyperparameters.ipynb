# Hyperameters of Logistic Regression in Scikit-Learn

1. **penalty**: This hyperparameter specifies the type of regularization to apply to the logistic regression model. Regularization helps prevent overfitting by adding a penalty term to the loss function. Common options include 'l1' (Lasso), 'l2' (Ridge), and 'elasticnet' (a combination of L1 and L2).



2. **dual**: This hyperparameter is a boolean that indicates whether to solve the dual or primal optimization problem. 

 

3. **tol**: This hyperparameter specifies the tolerance for stopping criteria. It determines when the optimization algorithm should stop iterating based on the change in the loss function.

4. **C**: This hyperparameter is the inverse of regularization strength. A smaller value of C indicates stronger regularization, while a larger value indicates weaker regularization. It controls the trade-off between fitting the training data well and keeping the model simple to avoid overfitting.

5. **fit_intercept**: This hyperparameter is a boolean that indicates whether to include an intercept term in the model. If set to True, the model will learn an intercept term; if False, it will not.

6. **class_weight**: This hyperparameter is used to assign weights to different classes in the dataset. It can be set to 'balanced' to automatically adjust weights inversely proportional to class frequencies

7. **solver**: This hyperparameter specifies the algorithm to use for optimization. Common options include 'liblinear', 'sag', 'saga', and 'lbfgs'. The choice of solver can affect the convergence and performance of the model.



8. **max_iter**: This hyperparameter specifies the maximum number of iterations for the optimization algorithm to converge. If the algorithm does not converge within this number of iterations, it will stop and return the best solution found so far.

9. **multi_class**: This hyperparameter specifies the strategy to use for multi-class classification. Common options include 'ovr' (one-vs-rest) and 'multinomial'. The choice of strategy can affect the performance of the model on multi-class datasets.

10. **verbose**: Training hote time agar aapko beech k stages mei result dekhna hai, then you can set verbose to 1. It will print the progress of the training process.

11. **warm_start**: This hyperparameter is a boolean that indicates whether to reuse the solution of the previous call to fit as initialization for the next call. If set to True, it can speed up convergence when fitting the model multiple times with different subsets of data.

12. **n_jobs**: This hyperparameter specifies the number of CPU cores to use for parallel processing. Setting it to -1 will use all available cores, which can speed up training on large datasets.

13. **l1_ratio**: This hyperparameter is used when the penalty is set to 'elasticnet'. It specifies the mixing parameter between L1 and L2 regularization. A value of 0 corresponds to L2 regularization, while a value of 1 corresponds to L1 regularization.
