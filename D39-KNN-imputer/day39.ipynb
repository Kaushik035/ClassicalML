# KNN Imputation

- Iska basic fanda hai ki you behave like your neighbours.
- So if any value is missing in a cell of a row, then we will look for the row that is most similar to it and then we will take the value of that cell from that row and put it in the missing cell.
- Nearest neighbour is found using euclidean distance.
- So if you take K nearest neighbours, then you will take the average of those K nearest neighbours and put it in the missing cell.
    - Ex. K= 3, then Value in the missing cell = (Value of 1st nearest neighbour + Value of 2nd nearest neighbour + Value of 3rd nearest neighbour) / 3

## But there are some problems with KNN imputation:
- We cant compute euclidean distance if there are missing values in the row.
- So we come up with concept of NAN_euclidean distance.




- Jahape vhi values missing hai, waha pe vhi values ko ignore kar denge aur baaki ke values ke basis pe distance calculate karenge.

- We are trying to fill the missing value in row 2,col 1 of the below pic



- So, for K = 2 , closest 2 rows are row 3 and row 4.
- So, value in the missing cell = (Value of 1st nearest neighbour + Value of 2nd nearest neighbour) / 2
- So, value in the missing cell = (5.19 + 4.9) / 2 = 5.045

## Advantages of KNN imputation:
1. More accurate than mean/median imputation.

## Disadvantages of KNN imputation:
1. Computationally expensive.
2. Not suitable for large datasets as while production we need to store the entire training dataset to find the nearest neighbours for the test data.

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
df.isnull().mean() * 100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

-  Parameters that can be tuned in KNN imputation:


- weight = uniform means that all the nearest neighbours will have equal weightage.
- weight = distance means that the nearest neighbour will have more weightage than the farthest neighbour. (we multiply the value of the nearest neighbour with 1/distance and then we take the average)


```python
knn = KNNImputer(n_neighbors=3,weights='distance')

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf,y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```
