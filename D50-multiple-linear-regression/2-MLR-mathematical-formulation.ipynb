## $\implies \hat{Y} = X\beta$

- So if we know the values of $\beta$ then we can easily calculate the predicted value of package for all the $n$ students by multiplying the data matrix $X$ with the parameter vector $\beta$

- Now let's see how to find the values of $\beta$ in multiple linear regression.

- Defining $Y$ and $e$ matrix:



- The objective loss function for multiple linear regression is in matrix form:







- The term underlined in red is equal, proof:

We want to show that:
$$
Y^T X \beta = (X\beta)^T Y
$$

Let us define:
$$
\hat{Y} = X\beta
$$

Then the two expressions become:
$$
Y^T X \beta = Y^T \hat{Y}
$$

and
$$
(X\beta)^T Y = \hat{Y}^T Y
$$

Now, note that both are scalars (since $(Y^T \hat{Y})$ is $(1 \times 1)$). and for $(1 \times1)$ matrices, the transpose does not change the value

Using the property:
$$
a^T b = b^T a
$$

we get:
$$
Y^T \hat{Y} = \hat{Y}^T Y
$$

Hence,
$$
Y^T X \beta = (X\beta)^T Y
$$

- So our loss function can be written as:



We want to compute:
$$
\frac{d}{d\beta} \left( \beta^T X^T X \beta \right)
$$

---

### Step 1: General Identity

For any matrix \(A\),
$$
\frac{d}{d\beta} \left( \beta^T A \beta \right) = (A + A^T)\beta
$$

---

### Step 2: Apply to our case

Here,
$$
A = X^T X
$$

So,
$$
\frac{d}{d\beta} \left( \beta^T X^T X \beta \right)
= (X^T X + (X^T X)^T)\beta
$$

---

### Step 3: Use transpose property

$$
(X^T X)^T = X^T X
$$

(since it is symmetric)

---

### Final Result

$$
\frac{d}{d\beta} \left( \beta^T X^T X \beta \right)
= 2 X^T X \beta
$$



- $(X^T \times X)^{-1}$ is a symmetric matrix so its transpose is itself:
$$
(X^T X)^{-1} = ((X^T X)^{-1})^T
$$
### Showing that $(X^T X)^{-1}$ is symmetric

A matrix \(A\) is symmetric if:
$$
A^T = A
$$

---

### Step 1: Take transpose

$$
(X^T X)^T
$$

Using the property:
$$
(AB)^T = B^T A^T
$$

we get:
$$
(X^T X)^T = X^T (X^T)^T
$$

---

### Step 2: Simplify

$$
(X^T)^T = X
$$

So,
$$
(X^T X)^T = X^T X
$$

---

### Conclusion

$$
(X^T X)^T = X^T X
\;\Rightarrow\;
X^T X \text{ is symmetric}
$$

---

## Inverse + Transpose Property

For any invertible matrix \(A\):
$$
(A^{-1})^T = (A^T)^{-1}
$$

---

### Apply to \(X^T X\)

Let:
$$
A = X^T X
$$

Then:
$$
\left[(X^T X)^{-1}\right]^T = \left[(X^T X)^T\right]^{-1}
$$

---

### Use symmetry

Since:
$$
(X^T X)^T = X^T X
$$

we get:
$$
\left[(X^T X)^{-1}\right]^T = (X^T X)^{-1}
$$

---

### Final Conclusion

$$
(X^T X)^{-1} \text{ is also symmetric}
$$

---

### Intuition

- \(X^T X\) is symmetric  
- Inverse of a symmetric matrix is also symmetric  
- So transpose has no effect:
$$
\left[(X^T X)^{-1}\right]^T = (X^T X)^{-1}
$$

## <span style="color:orange">Why do we use Gradient Descent when a closed-form solution</span> $\beta = (X^T X)^{-1} X^T Y$ <span style="color:orange">exists?</span>

- Because of the computational cost of calculating the inverse of a matrix, which is $O(m^3)$ where $m$ is the number of features. For large datasets with many features, this can be very expensive. Gradient Descent, on the other hand, has a computational cost of $O(m)$ per iteration, making it more efficient for large datasets.

- So Gradient Descent is a faster technique for finding the optimal parameters in multiple linear regression, especially when dealing with large datasets.
