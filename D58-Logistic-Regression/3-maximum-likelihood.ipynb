# Maximum Likelihood 

- It is the method we will use to find the cost function for logistic regression



- we find $\hat{y} = \sigma(z)$, and $\sigma(z)$ is the probability that the instance belongs to the positive class (green class), so probability of instance belonging to negative class (red class) is $1 - P(G)$

- Maximum likelihood batata hai ki, saare points k probability ko multiply kar do, aur jis model ke liye ye product maximum hoga, wahi best model hoga

- #### Note: Multiply krte time green point k liye hume $P(G)$ ko multiply karna hai, aur red point k liye $P(R)1 = 1 - P(G)$ ko multiply karna hai, doesn't matter ki point kis side hai



- we can see in the above pic that model 2 has higher likelihood than model 1, because it classifies the data better than model 1

- ### So, we need to find the coefficients of the line that maximize the likelihood function

- But here the problem is that, here we multipled ony 4 probabilities, still we got a very small number, imagine if we have 1000 points, then we will get a very small number, which is not good for comparison

- So , we need to find a way to convert this product into sum

- So, we take log of the likelihood function, because log of a product is equal to sum of logs, so we can convert the product into sum



- But here problem is that ,as the number inside the log is between 0 and 1, so log of that number will be negative, so we need to take negative log of the likelihood function, so that we can get a positive number



- ### <span style="color:orange">Cross Entropy = negative log of likelihood function </span>

- ### So, we need to find the coefficients of the line that **minimize the cross entropy** function (because we want to maximize the likelihood function, so we need to minimize the negative log of the likelihood function, as log(0.1) > log(0.8), so we need to minimize the cross entropy function)



- We directly can't use this formula in the above pic, as for green points we need to use $ \hat{y} = P(G) $ and for red points we need to use $P(R) = 1 - \hat{y} $, so we need to use the below formula for cross entropy



- So, if a point is actually green, then we will use $-log(\hat{y})$ for that point, and if a point is actually red, then we will use $-log(1 - \hat{y})$ for that point

- So the final Loss function for logistic regression is given by the below formula:

### $$L(w) = -\frac{1}{N} \sum_{i=1}^{N} [y_i log(\hat{y_i}) + (1 - y_i) log(1 - \hat{y_i})]$$

- Where $N$ is the number of points, $y_i$ is the actual label of the point (1 for green and 0 for red), and $\hat{y_i}$ is the predicted probability of the point belonging to the positive class (green class)

- This loss function is also known as the **log loss error** or **binary cross entropy loss** function

- So, we need to find the coefficients of the line that minimize this loss function, and that will be our best model that classifies the data well

- But there is no closed form solution for this loss function, so we need to use  **gradient descent** to find the coefficients of the line that minimize this loss function
