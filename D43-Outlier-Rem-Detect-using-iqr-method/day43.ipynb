# Outlier Removal and Detection using IQR Method



- The wiskers is 1.5 times the IQR (Interquartile Range) from the first and third quartiles.
- Any data point outside of the wiskers is considered an outlier.

### Step 0: See if the data is skewed
### Step 1: Q1,Q3 and IQR Calculation
### Step 2: Calculate the wiskers
### Step 3: Trimming/Capping (value outside the wiskers)



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.kdeplot(df['cgpa'])
# sns.histplot(df['cgpa'], kde=True)

plt.subplot(1,2,2)
sns.kdeplot(df['placement_exam_marks'])

plt.show()
```

```python
df['cgpa'].skew() # close to zero, so it is normally distributed
```

```python
df['placement_exam_marks'].skew() # right skewed
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
# Finding the IQR
percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
```

```python
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

## Finding Outliers

```python
df[df['placement_exam_marks'] > upper_limit]
```

```python
df[df['placement_exam_marks'] < lower_limit]
```

## Trimming

```python
new_df = df[df['placement_exam_marks'] < upper_limit]
```

```python
new_df.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```

## Capping

```python
new_df_cap = df.copy()

new_df_cap['placement_exam_marks'] = np.where(
    new_df_cap['placement_exam_marks'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['placement_exam_marks'] < lower_limit,
        lower_limit,
        new_df_cap['placement_exam_marks']
    )
)
```

```python
new_df_cap.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['placement_exam_marks'])

plt.show()
```
