# Curse of Dimensionality

- Columns <-> Dimensions <-> Features <-> Attributes <-> Variables all refer to the same thing in the context of datasets.

- For a ML model, there is a "sweet spot" for the number of features. Too few features may lead to underfitting, while too many features can lead to overfitting and increased computational complexity.

### Examples of High-Dimensional Data
- **Image Data**: A 100x100 pixel image has 10,000 features (one for each pixel).
- **Text Data**: Using a bag-of-words model, a document can have thousands of features corresponding to the presence of different words.

# What is the problem with high-dimensional data?

- **Sparsity**: As the number of dimensions increases, the data becomes sparse. This means that the data points (rows) are spread out in the feature space, making it difficult for models to find meaningful patterns.

- Leading to decreased performance of ML models and increased computational cost.

## Example: You lost your keys in the school, and have the following clues:
- You lost it in way to Gate to your school building. (1D)
- you lost it in the school playground. (2D)
- You lost it in the school building. (3D)

- In 1D, you can easily find your keys by searching along the path to the gate.
- In 2D, you have to search the entire playground, which is more time-consuming.
- In 3D, you have to search the entire school building, which is even more time-consuming and difficult.


# What is the solution to the Curse of Dimensionality?

## <span style="color:orange">Dimensionality Reduction</span>:
- Reduce the number of columns (features)
- Two main approaches:
  - **Feature Selection**: Select a subset of the original features based on some criteria 
    1. **Forward Selection**: Start with no features and add one feature at a time based on performance improvement.
    2. **Backward Elimination**: Start with all features and remove one feature at a time based on performance degradation.
  - **Feature Extraction**: Create new features by transforming the original features
    1. **Principal Component Analysis (PCA)**: Taking linear combinations of the original features to create new features that capture the most variance in the data.
    2. **LDA**: Linear Discriminant Analysis, which is a supervised method that finds a linear combination of features that best separates the classes.
    3. **t-SNE**: t-Distributed Stochastic Neighbor Embedding, which is a non-linear method that is particularly good for visualizing high-dimensional data in 2D or 3D.
