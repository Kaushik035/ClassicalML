# 2. Numerical Data

- Numerical data mei main hota hai ki whats mean, median, mode, standard deviation, min, max, range, percentiles etc.

- But usse vhi pehle we can visualize the numerical data a categorical way, here we convert range of numerical data into bins.. to visualise the distribution of numerical data...And for that we use histogram

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=11)
```

## b. Distplot

- Displot is kind of improvement over histplot as it combines histogram and KDE(Kernel density estimate) plot together.

```python
sns.distplot(df['Age'])
```

- The KDE (blue line) is the probablity density function
- So agar hum kisi bande ko randomly select karte hai to uske age ke probability kya hai wo hume KDE batata hai.
- It also tells us wether there is any skewness in the data or not.

## c. Boxplot

- The work of the boxplot is to show us the 5 number summary of the numerical data along with outliers if any.

- In the above boxplot:
    - The line in the middle is median (50th percentile, this means 50% data is below this line and 50% data is above this line)
    - The box is from Q1 to Q3 (25th percentile to 75th percentile)
    -Q1 means 25% data is below this point and 75% data is above this point
    - Q3 means 75% data is below this point and 25% data is above this point
    - The length of the box is called IQR (Interquartile range) = Q3 - Q1
    - The whiskers (lines extending from the box) goes upto 1.5*IQR from Q1 and Q3, which helps us to identify outliers that don't follow the general distribution of the data.
    - The dots are outliers

- And isko hum generally tb use krte hai jab humae data mei noise ya outliers dikh rahe hote hai.

```python
sns.boxplot(x=df['Fare'])
```
