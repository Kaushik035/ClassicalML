## 1. Feature Transformation
- We have the data column, but we change the  data column to a different format for better model performance, which is called feature transformation.

### 1.1 Missing Value Imputation

- The sklearn model cannot handle missing values, so we need to fill in the missing values with some values or remove the rows/columns with missing values, which is called missing value imputation.



### 1.2 Handling Categorical Values

- Our sklearn model cannot handle categorical values, so we need to convert them into numerical values.



### 1.3 Outlier Detection

- Outliers are data points that are significantly different from other data points. They can affect the performance of the model, so we need to detect and handle them.



### 1.4 Feature Scaling
- Feature scaling is the process of normalizing the range of features. For example in the pic, suppose we calculade euclidean distance between two data points, the feature with larger range will dominate the distance calculation, which may lead to poor model performance. Therefore, we need to scale the features to a similar range.(Here salary has a larger range than age, so we need to scale salary to a similar range as age)



## 2. Feature Construction
- We create a completely new column manually based on the domain knowledge
- Like here we combined the siblings and parents column to create a new column called family size, which categorizes the family size, which may be more useful for the model than the original columns.


## 3. Feature Selection
- Humare paas bohot saare features/input columns hain, or hum unme se kuch features ko select karte hain jo model ke liye useful hain, or baaki features ko drop kar dete hain, jise feature selection kehte hain.
- Like in the below pic, we can eliminate the boundary side pixels and take ony the center pixels to classify the image, which can reduce the number of features and improve the model performance.



## 4. Feature Extraction
- Given features/ columns mei se hum completely new features create karte hain programmatically (not manually-> wo feature construction hota hai), jise feature extraction kehte hain.
