# Recall



- Out of both the cancer detection models, which one will you choose?

    - So, the FN is more costly than the FP, so we will choose model A, as it has less FN

- This thing is called Recall : **What proportion of actual positive are predicted positive?**







- #### So, when accuracy is not a good metric, we can use precision and recall to evaluate the model performance. But we have to choose the right metric based on the problem we are trying to solve.
- #### If we want to minimize the FP, we will use precision, and if we want to minimize the FN, we will use recall.

# F1 Score

- Sometimes we can't determine which error is more costly, like classification wether the picture is of a cat or a dog

- So here we have to balance both the precision and recall (**as there is a trade off between them**), so we will use F1 score, which is the harmonic mean of precision and recall

- Harmonic mean is used because it gives more weight to the lower value, so if either precision or recall is low, the F1 score will be low, which means it penalizes the model for the low value (unlike the arithmetic mean, which can be high if one of the values is high as it doesn't give more weight to the lower value and wants to balance both the values)



$$
F1 = 2 \times \frac{\text{precision} \times \text{recall}}{\text{precision} + \text{recall}}
$$

- In the above pic you can see for the case B, P =60, R = 100, so F1 score will be 75, and for the case A (The F1 score is more towards P, as it gives more weight to precision {lower value}), P = 80, R = 80, so F1 score will be 80, so we will choose case A as it has higher F1 score, which means it has a better balance between precision and recall.

# How to calculate Precision, Recall and F1 score for a multi class classification problem?

- For binary calssification, we mostly focus on the positive class (1). For eg. Is a email spam (1) or not (0), do you have cancer (1) or not (0)

- So, the precision, recall and F1 score that it shows is for the positive class (1) by default.

- But behind the scenes, it is calculating the precision, recall and F1 score for each class (0 and 1) 





- So, in binary classification we take the precision, recall and F1 score of the positive class (1) as the final metric

# Calculating Precision for multi class classification



### Precision is the proportion of predicted positive that are actually positive

$$
\text{Precision for dog} = \frac{\text{Actual kitne dogs}}{\text{Number of predicted dogs}}
$$
and similarly for cat and rabbit



- ## Now we have 2 ways to calculate the combined precision, recall and F1 score for all the classes

    - Macro Precision : It is the average of the precision, recall and F1 score of all the classes. It gives equal weight to all the classes, so it is not affected by the class imbalance.

    - Weighted Precision : It is the average of the precision, recall and F1 score of all the classes.

### When to use Macro Precision and when to use Weighted Precision?

- If we have a balanced dataset, we can use Macro Precision, as it gives equal weight to all the classes.

- On the other hand, if we have an imbalanced dataset, we can use Weighted Precision, as it gives more weight to the majority class.



# Calculating Recall for multi class classification

- Recall is the proportion of actual positive that are predicted positive.

$$
\text{Recall for dog} = \frac{\text{Predicted number of dogs}}{\text{Actual number of dogs}}
$$



# Calculating F1 score for multi class classification



- Similar as we calculated Precision and Recall, we can calculate F1 score for each class and then take the average of them to get the combined F1 score for all the classes. We can use either Macro F1 score or Weighted F1 score based on the class imbalance in the dataset.
