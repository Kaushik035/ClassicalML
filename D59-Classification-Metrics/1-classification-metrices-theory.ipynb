# How much accuracy is good?

- **It depends on the problem we are solving**

- For example, if we are solving a problem of detecting cancer, we want to have a very high accuracy (close to 100%) because we don't want to miss any cancer cases. On the other hand, if we are solving a problem of detecting spam emails, we can tolerate a lower accuracy (say 90%) because we can afford to miss some spam emails.

# The Problem with Accuracy Metrix

- The Accuracy score doesn't tell ki konsi type ki galti ho rahi hai.

- Fo example we made a model to predict if an email is spam or not (binary classification), and we got an accuracy of 90%. This means that our model is correct 90% and incorrect 10% of the time.

- But there are two types of errors that our model can make:

    - False Positive: when the model predicts that an email is spam when it's not (Type I error)
    - False Negative: when the model predicts that an email is not spam when it is (Type II error)

- So, in that 10% of the time that our model is incorrect, we don't know how many of those are False Positives and how many are False Negatives. This is a problem because in some cases, one type of error might be more costly than the other. For example, in the case of detecting cancer, a False Negative (missing a cancer case) is much more costly than a False Positive (predicting cancer when it's not there).

- So, we need to look at other metrices that can give us more insights into the types of errors our model is making, such a matrix is **Confusion Matrix**.

# Confusion Matrix



- If actually its a positive case and we predicted it as positive, it's a True Positive (TP)
- If actually its a negative case and we predicted it as negative, it's a True Negative (TN)
- If actually its a negative case and we predicted it as positive, it's a False Positive (FP)
- If actually its a positive case and we predicted it as negative, it's a False Negative (FN)

- ### So, the positive/negative comes from the prediction, and the true/false comes from the actual label, that wether the prediction is correct or not.

- ### Cheat code:
    - 1st word : True/False (By looking at the actual label)
    - 2nd word : Positive/Negative (By looking at the prediction)

# Accuracy from Confusion Matrix

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

# Types of Errors

- ## 1. Type I Error (False Positive)
    - This is when we predict a positive case when it's actually negative (FP).
    - This is also called a False Alarm.


- ## 2. Type II Error (False Negative)
    - This is when we predict a negative case when it's actually positive (FN).
    - This is also called a Miss.

# When is Accuracy misleading?

1. **Imbalanced Datasets**: When the classes are imbalanced, accuracy can be misleading. For example, if we have a dataset with 95% of the samples belonging to one class and only 5% belonging to the other class, a model that always predicts the majority class will have an accuracy of 95%, which seems good but is actually not useful.

    - Eg. To detect in airport securitythat a person is a terrorist or not, and lets say our data set have 10000 passenger, and out of them only 1 is terrorist, so even if you fit a model that always predicts that passeges is not a terrorist , still it will mistake only once (Though it will have a high accuracy)



- So, we cant rely on accuracy on this type of case, so we study metrices like **Precision** and **Recall**.
